## Cần chuẩn hóa đặc trưng, vì:
1. Khi các feature có giá trị chênh lệch khá lớn. Ví dụ
- Feature 1: [300, 2000]
- Feature 2: [0, 5]

Điều này làm cho các trục có đơn vị và thang đo khác nhau, dẫn đến:
-  Hàm $J(w, b)$ - contour plot - bị kéo dài thành ellipse hẹp và nghiêng 
-  Gradient Descent khi cập nhật tham số $w$ sẽ đi zig-zag (độ dốc ở mỗi trục khác nhau)

&rarr; Hội tụ chậm, cần learning rate nhỏ 

2. Khi chuẩn hóa về cùng khoảng giá trị, e.g: [0, 1), contour plot *tròn* hơn

&rarr; Gradient Descent đi đến Global Minimum nhanh hơn, gần như trên đường thẳng 

### Trước khi scaling
- Contour plot: các đường elipse, hàm $J(w_1, w_2)$ thay đổi nhanh hơn theo 1 hướng (dốc đứng) và chậm theo hướng còn lại
    &rarr; nếu $\alpha$ - learning rate lớn: nó sẽ nhảy qua nhảy lại - hình zig-zag - điểm Global Minimum
