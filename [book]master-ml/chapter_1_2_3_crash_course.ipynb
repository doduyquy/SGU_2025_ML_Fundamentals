```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, norm, multivariate_normal, beta, dirichlet, randint

# 1. Bernoulli
plt.figure(figsize=(15, 10))

plt.subplot(231)
p = 0.7
x = [0, 1]
pmf = [1-p, p]
plt.bar(x, pmf)
plt.title(f"Bernoulli(p={p})")

# 2. Categorical (K=3)
plt.subplot(232)
probs = [0.1, 0.3, 0.6]
x = np.arange(len(probs))
plt.bar(x, probs)
plt.title(f"Categorical(p={probs})")

# 3. Normal (1D)
plt.subplot(233)
x = np.linspace(-5, 5, 200)
plt.plot(x, norm.pdf(x, 0, 1), label="=0, =1")
plt.plot(x, norm.pdf(x, 0, 2), label="=0, =2")
plt.plot(x, norm.pdf(x, 2, 1), label="=2, =1")
plt.legend()
plt.title("Normal Distribution")

# 4. Multivariate Normal (2D contours)
plt.subplot(234)
mean = [0, 0]
cov1 = [[1, 0], [0, 1]]   # c lp
cov2 = [[1, 0.8], [0.8, 1]]  # tng quan
x, y = np.mgrid[-3:3:.05, -3:3:.05]
pos = np.dstack((x, y))
rv1 = multivariate_normal(mean, cov1)
rv2 = multivariate_normal(mean, cov2)
plt.contour(x, y, rv1.pdf(pos), levels=5, colors="blue", linestyles="dashed")
plt.contour(x, y, rv2.pdf(pos), levels=5, colors="red")
plt.title("Multivariate Normal")

# 5. Beta
plt.subplot(235)
x = np.linspace(0, 1, 200)
plt.plot(x, beta.pdf(x, 0.5, 0.5), label="=0.5, =0.5")
plt.plot(x, beta.pdf(x, 2, 2), label="=2, =2")
plt.plot(x, beta.pdf(x, 5, 1), label="=5, =1")
plt.legend()
plt.title("Beta Distribution")

# 6. Dirichlet (K=3) -> minh ha mu ngu nhin
plt.subplot(236)
alpha = [2, 2, 2]
samples = dirichlet.rvs(alpha, size=500)
plt.scatter(samples[:,0], samples[:,1], alpha=0.5, s=10)
plt.title("Dirichlet(=[2,2,2]) samples (projected)")

plt.tight_layout()
plt.show()
```
