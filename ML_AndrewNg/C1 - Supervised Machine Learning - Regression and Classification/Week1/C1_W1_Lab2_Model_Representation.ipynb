```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print(f"x_train =  {x_train}")
print(f"y_train =  {y_train}")
```

```python
# m is the number of training examples
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]    # get the number of rows
print(f"Number of training examples is: {m}")
```

```python
# show all the data
for i in range(len(x_train)):
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='o', color='green', label="Data points")
# Set the title
plt.title("House Prices")
# Set the x-axis label
plt.xlabel("Size (1000 sqft)")
# Set the y-axis label
plt.ylabel("Price (in 1000s of dollars)")
plt.legend()    # show the label in scatter/plot
plt.show()
```

```python
# set an arbitrary(random) weight and bias for drawing the regression line
w = 100
b = 21
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Compute the prediction of a linear model
    Args:
        x (ndarray (m, )): data, m examples
        w, b (scalar)    : model parameters
    Returns
        y (ndarray (m,)):  target values
    """
    m = x.shape[0]  # m = 2
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb

# compute_model_output(x_train, w, b)
```

```python
# Plot function
def plot_data(f_wb):
    # Plot our model prediction
    plt.plot(x_train, f_wb, color='blue', label="Our Prediction") 

    # Plot the data points
    plt.scatter(x_train, y_train, marker='x', color='red', label='Actual Values')

    # Set the title
    plt.title("House Prices")
    # Set the x-axis label
    plt.xlabel("Size (1000 sqft)")
    # Set the x-axis label
    plt.ylabel("Price (in 1000s of dollars)")
    plt.legend()
    plt.show()
```

```python
plot_data(compute_model_output(x_train, w, b))
```

- It not fit. To improve it &rarr; experiment with w & b

```python
# The right w & b
w = 200
b = 100
plot_data(compute_model_output(x_train, w, b))
```

```python
# Thus, let's get an prediction from our model on some arbitrary price (x)
x_i = 1.3
cost_sqft = w * x_i + b

print(f"Cost of {(x_i*1000):.0f} sqft: ${cost_sqft}")
```
