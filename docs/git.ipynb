# SUBMODULE IN YOUR GIT

## Option 1: Save submodule like a normal directory

1. Delete .git of submodule

    `rm -rf ML_AndrewNg/Template/Machine-Learning-Specialization-Coursera/.git`
    <!-- ```console
    `rm -rf ML_AndrewNg/Template/Machine-Learning-Specialization-Coursera/.git`
    ``` -->

2. Add all file in submodule to its parent git

    `git add .`

3. Push to git again

    `git push`

## Option 2: Keep it as a submodule

1. Remove git of submodule (only git, not file/dir)

    `git rm --cached ML_AndrewNg/Template/Machine-Learning-Specialization-Coursera`

2. Add again with submodule (type)

    `git submodule add <url> ML_AndrewNg/Template/Machine-Learning-Specialization-Coursera`

3. Commit change

    `git commit -m "Add Machine-Learning-Specialization-Coursera as submodule"`



# CONFIG MERGE | REBASE | FAST-FORWARD

When you pull from git, having an conflict, but it don't show anything to merge,..

+ Config your pull:
    - Option 1: Merge 

        `git config pull.rebase false`

    - Option 2: Rebase

        `git config pull.rebase true`

    - Option 3: Fast-forward only

        `git config pull.ff only`
